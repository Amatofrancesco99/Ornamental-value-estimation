```python
def row_to_json(row):
    
    """
    Converts a row of this dataset into a json object with a variable number of fields.
    """
    li = []
    res = "{"
    
    for key in row.keys():
        
        val = row[key]
        
        if val == "X":
            val = 'true'
            
        try:
            int(val)
            float(val)
        except:
            if val!='true':
                val = f"\"{val}\""
        
        li.append((key, val))
    
    for tup in li:
        res+=f"\"{tup[0]}\" : {tup[1]}, "
    
    res = res.strip(", ")
    res+="}"
    return res
```

```python
def dataframe_to_json_array(df):
    """
    Convert a dataframe into a jsonarray.
    Each row of the df is treated as a json object.
    """
    json_objects = df.apply(lambda x : row_to_json(x), axis=1)
    json_array = "{ \"Plants\" : [ "
    for obj in json_objects:
        json_array+=obj+", "
    json_array = json_array.strip(", ")
    json_array+=" ]}"
    return json_array
```

## Obtain ornamental score from plants
<br>

We can see, from the *with_scores_df* that the ornamental score of a single species is a function of this variables:
* 4030
* 2330 + 6210
* 6210
* VISCOSITA'
* PERIODO FORNITURA
* PRESENZA MERCATO
* AROMATICA' 
* IMPOLLINAZIONE
* ALLERGENICITA'
* TOSSICITA'

```python
def get_score_sum(singular_species):
    overall = 0
    if ((singular_species[4030] == 'True')
    or (singular_species["2330+6210(acid)"] == 'True')
    or (singular_species["6210(calc)"] == 'True')
    or (singular_species["VISCOSITA'"] == 1)
    or (singular_species["PRESENZA MERCATO"] != "nan")
    or (singular_species["AROMATICA"] == 1)
    or (singular_species["IMPOLLINAZIONE"] == 1)
    or (singular_species["ALLERGENICITA'"] == 1)
    or (singular_species["TOSSICITA'"] == 1)
    or (singular_species["AROMATICA"] == 1)):
        overall += 1
    # IMPROVE THE FOLLOWING SCRIPT PART
    if (singular_species["PERIODO_FIORITURA"] == ""):
        overall += 1
    elif (singular_species["PERIODO_FIORITURA"] == ""): 
        overall += 2
    elif (singular_species["PERIODO_FIORITURA"] == ""):
        overall += 3
    elif (singular_species["PERIODO_FIORITURA"] == ""):
        overall += 4
    return overall
```

```python
def get_ornamental_index_value(singular_species):
    score_sum = get_score_sum(singular_species)
    if (score_sum < 6):
        return 1
    elif (score_sum >= 6) and (score_sum < 10):
        return 2
    elif (score_sum >= 10):
        return 3
```
