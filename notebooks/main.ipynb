## Import

```python
## Import libraries
import pandas as pd
import json
from ipynb.fs.full.utilities import dataframe_to_json_array, get_score_sum, get_ornamental_index_value, store_into_mongodb_istance

# Import dataset:
raw = "http://csu.unipv.it/wp-content/uploads/2022/01/Database_ornamentali_grezzo.xls"
```

### Converting the .xls to dataframe

```python
# Get dataframes starting from existing paths
raw_df = pd.read_excel(raw)
```

```python
raw_df
```

## Obtain ornamental score from plants

### Test: compute the ornamental value for a single row/record/plant:

```python
"""
Remember that "get_score_sum" & "get_ornamental_index_value" functions are contained into the 
other .ipynb file, named as "utilities"
"""

print('SCORE SUM: ' + str(get_score_sum(raw_df.loc[0])))
print('ORNAMENTAL VALUE: ' + str(get_ornamental_index_value(raw_df.loc[0])))
```

### Goal: do it for all of the plants:

```python
""" SCORE SUM FOR EACH PLANT """
raw_df.apply(get_score_sum, axis=1)
```

```python
""" ORNAMENTAL VALUE FOR EACH PLANT """
raw_df.apply(get_ornamental_index_value, axis=1)
```

## Generate the json file

We want to add our figures of merit to the single plant and then create a single json file containing all this information.

```python
# Add the sum score and ornamental value to each plant
sum_of_scores = list(raw_df.apply(get_score_sum, axis=1))
all_ornamental_values = list(raw_df.apply(get_ornamental_index_value, axis=1))

raw_df["SOMMA DEI PUNTEGGI"] = sum_of_scores
raw_df["INDICE DI ORNAMENTALITA'"] = all_ornamental_values
```

```python
""" SHOW THE NEW DATAFRAME WITH FIGURES OF MERIT """
raw_df
```

```python
"""
Remember that "dataframe_to_json_array" function is contained into the other .ipynb file, named
as "utilities"
"""

json_array = dataframe_to_json_array(raw_df)
```

## Store the obtained json into our mongo-db istance

```python
"""
Remember that "store_into_mongodb_istance" function is contained into the other .ipynb file, named
as "utilities"
"""

store_into_mongodb_istance(json.loads(json_array))
```
