## This notebook deals only with the **polished version** of the dataset, converting it from xls into json format:

```python
with_scores = "http://csu.unipv.it/wp-content/uploads/2022/01/Database_ornamentali_201006-con-punteggi.xls"
```

```python
# Get dataframes starting from existing paths
df = pd.read_excel(with_scores)
```

```python
df
```

```python
def row_to_json(row):
    
    """
    Converts a row of this dataset into a json object with a variable number of fields.
    """
    li = []
    res = "{"
    
    for key in row.keys():
        
        val = row[key]
        
        if val == "X":
            val = 'true'
            
        try:
            int(val)
            float(val)
        except:
            if val!='true':
                val = f"\"{val}\""
        
        li.append((key, val))
    
    for tup in li:
        res+=f"\"{tup[0]}\" : {tup[1]}, "
    
    res = res.strip(", ")
    res+="}"
    return res
```

```python
def dataframe_to_json_array(df):
    """
    Convert a dataframe into a jsonarray.
    Each row of the df is treated as a json object.
    """
    json_objects = df.apply(lambda x : row_to_json(x), axis=1)
    json_array = "{ \"Plants\" : [ "
    for obj in json_objects:
        json_array+=obj+", "
    json_array = json_array.strip(", ")
    json_array+=" ]}"
    return json_array
```

## Convert the dataframe into a json-array

```python
json_array = dataframe_to_json_array(df)
```
